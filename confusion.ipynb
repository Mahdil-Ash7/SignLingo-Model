```python
# 05_confusion_analysis.py
# Run after 02_train_model.py to find confused sign pairs

import numpy as np
import json
import pickle
from sklearn.metrics import confusion_matrix
import tensorflow as tf

MODEL_PATH   = "bim_model/handface_pose_cnn_gru.h5"
LABEL_PATH   = "bim_model/labels.json"
ENCODER_PATH = "bim_model/label_encoder.pkl"

model = tf.keras.models.load_model(MODEL_PATH, compile=False)
with open(LABEL_PATH)   as f: labels  = json.load(f)
with open(ENCODER_PATH, 'rb') as f: le = pickle.load(f)

# Load test set — rerun the same split as training
from training_model import load_dataset, normalize_sequence
from sklearn.model_selection import train_test_split

X, y = load_dataset("dataset_desktop")
y_enc = le.transform(y)
_, X_test, _, y_test = train_test_split(
    X, y_enc, test_size=0.2, random_state=42, stratify=y_enc)

y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)
cm     = confusion_matrix(y_test, y_pred)

print("\n🔍 Confused pairs (>10% confusion rate):")
print(f"{'Sign A':<12} {'Sign B':<12} {'A→B%':>8} {'B→A%':>8}")
print("─" * 44)

for i, label_a in enumerate(labels):
    row_total = cm[i].sum()
    if row_total == 0: continue
    for j, label_b in enumerate(labels):
        if i == j: continue
        ab_rate = cm[i, j] / row_total * 100
        ba_rate = cm[j, i] / cm[j].sum() * 100 if cm[j].sum() > 0 else 0
        if ab_rate > 10 or ba_rate > 10:
            print(f"{label_a:<12} {label_b:<12} {ab_rate:>7.1f}% {ba_rate:>7.1f}%")
```
